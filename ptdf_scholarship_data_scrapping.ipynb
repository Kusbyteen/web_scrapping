```python
#importing libraries
import requests as re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://pid.ptdf.gov.ng/pid/pid/scholarship/?page=1"
response = re.get(url)
print(response)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
table_data = soup.find("table")
table_data
```

```python
# Collecting Ddata
l = []
table_data = soup.find("table")
for index in range(1, 922):
    for row in table_data.find_all("tr"):    
        # Find all data for each column
        columns = row.find_all('td')
        data = [cols.text.strip() for cols in columns]
        l.append(data)
    next_url = f"https://pid.ptdf.gov.ng/pid/pid/scholarship/?page={str(1 + index)}"
    url = next_url    
    print(url)
```

```python
column_headers = []
for col in table_data.find_all("th"):
    col_heads = col.text.strip().replace(' ', '_')
    column_headers.append(col_heads)
column_headers
```

```python
df = pd.DataFrame(l, columns=column_headers)
```
